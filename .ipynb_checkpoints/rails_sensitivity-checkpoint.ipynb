```python
import torch
import torch.nn as nn
from collections import deque
import numpy as np
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from models.VGG import VGG
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# local imports
from rails import RAILS
from aise import AISE
from pgd import PGD
```

```python
%matplotlib inline
```

```python
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(DEVICE)
```

```python
ROOT = "./datasets"
TRANSFORM = transforms.ToTensor()

train_data = datasets.CIFAR10(root=ROOT, 
                              download=True, 
                              train=True, 
                              transform=TRANSFORM)

test_data = datasets.CIFAR10(root=ROOT, 
                             download=True, 
                             train=False, 
                             transform=TRANSFORM)

# trainning data
x_train = (torch.FloatTensor(train_data.data[:50000])/255).permute(0,3,1,2)
y_train = torch.LongTensor(train_data.targets[:50000])


"""
baseline VGG model
"""

model = VGG()
model.load_state_dict(torch.load(
    "./model_weights/cifar_vgg16.pt", map_location=DEVICE
)['state_dict'])
model.eval()


"""
PGD attack
"""
pgd = PGD(eps=8/255,
          step=2/255,
          max_iter=10, 
          batch_size=256)


# evaluate the models and the attack
x_batch = (torch.FloatTensor(test_data.data[:1000])/255).permute(0,3,1,2)
y_batch = torch.LongTensor(test_data.targets[:1000])

x_adv = pgd.generate(model, x_batch, y_batch, device=DEVICE)

pred_clean = model(x_batch.to(DEVICE))[-1].max(dim=1)[1].detach().cpu()
clean_acc = (pred_clean == y_batch).float().mean().item()
print("Clean accuracy is {}.".format(clean_acc))

pred_adv = model(x_adv.to(DEVICE))[-1].max(dim=1)[1].detach().cpu()
adv_acc = (pred_adv == y_batch).float().mean().item()
print("Adversarial accuracy (CNN) is {}.".format(adv_acc))
```

```python
"""
rails with variable n_neighbors parameter 
"""

result_rows = []

low = 2
high = 20
by = 5
steps = int((high - low)/by)

for n in np.linspace(low, high, steps):
    CIFAR_CONFIGS = {
        "start_layer": 1,
        "n_class": 10,
        'aise_params' : [
            {"n_neighbors" : int(n), # variable 
             "n_population" : 10 * int(n),
             "hidden_layer": 2, 
             "sampling_temperature": 1, 
             "max_generation": 10, 
             "mut_range": (.005, .015)}
        ]
    }
    
    print(f"trainning with: n_neighbors = {int(n)}, n_population = {10 * int(n)} ")
    rails = RAILS(model, CIFAR_CONFIGS, x_train, y_train)
    
    pred_rails = rails.predict(x_batch.to(DEVICE)).argmax(axis=1)
    acc_rails = (pred_rails == y_batch.numpy()).astype("float").mean().item()
    
    pred_adv_rails = rails.predict(x_adv.to(DEVICE)).argmax(axis=1)
    adv_acc_rails = (pred_adv_rails == y_batch.numpy()).astype("float").mean().item()
    
    row = {
        'RAILS_clean_acc' : acc_rails,
        'RAILS_adversarial_acc' : adv_acc_rails,
        'n_neighbors' : int(n),
        'n_population' : 10 * int(n),
    }
    
    result_rows.append(row)
    
results = pd.DataFrame(result_rows)
results
```

```python
"""
simple plot for the above
"""

matplotlib.rcParams['figure.dpi'] = 150

plt.plot(results['n_neighbors'],
         results['RAILS_clean_acc'],
         c='black', 
         label="Clean Accuracy")

plt.plot(results['n_neighbors'],
         results['RAILS_adversarial_acc'],
         c='black', 
         ls=":",
         label="Adversarial Accuracy")
         
plt.title("Changes to Accuracy Based on N Neighbors")
plt.xlabel("N Neighbors")
plt.ylabel("RAILS Accuracy")
plt.legend()

outpath = f"results/n_neighbors.png"
plt.savefig(outpath, bbox_inches="tight")
```

```python
"""
does mutation rate affect the performance for large populations
"""

result_rows = []

N_NEIGHBORS = 3
N_POPULATION = 4

for n in np.linspace(0.001, 0.05, 3):
    for _n in np.linspace(0.001, 0.005, 3):
        mut_range = (n, n + _n)
        
        n_pop = N_POPULATION * N_NEIGHBORS * 10
    
        CIFAR_CONFIGS = {
            "start_layer": 1,
            "n_class": 10,
            'aise_params' : [
                {"n_neighbors" : N_NEIGHBORS, 
                 "n_population" : n_pop, # variable 
                 "hidden_layer": 2, 
                 "sampling_temperature": 1, 
                 "max_generation": 10, 
                 "mut_range": mut_range}
            ]
        }

        print(f"trainning with: mutation range = ({mut_range[0]:.3f} - {mut_range[1]:.3f})")
        rails = RAILS(model, CIFAR_CONFIGS, x_train, y_train)

        pred_rails = rails.predict(x_batch.to(DEVICE)).argmax(axis=1)
        acc_rails = (pred_rails == y_batch.numpy()).astype("float").mean().item()

        pred_adv_rails = rails.predict(x_adv.to(DEVICE)).argmax(axis=1)
        adv_acc_rails = (pred_adv_rails == y_batch.numpy()).astype("float").mean().item()

        row = {
            'RAILS_clean_acc' : acc_rails,
            'RAILS_adversarial_acc' : adv_acc_rails,
            'n_neighbors' : N_NEIGHBORS,
            'n_population' : n_pop,
            'mutation_range_L' : mut_range[0],
            'mutation_range_U' : mut_range[1],
        }

        result_rows.append(row)
    
results = pd.DataFrame(result_rows)
results
```

```python
matplotlib.rcParams['figure.dpi'] = 150

results['mutation_rate_centroid'] = (results['mutation_range_L'] + results['mutation_range_U'])/2

plt.plot(results['mutation_rate_centroid'],
         results['RAILS_clean_acc'],
         c='black', 
         label="Clean Accuracy")

plt.plot(results['mutation_rate_centroid'],
         results['RAILS_adversarial_acc'],
         c='black',
         ls=":",
         label="Adversarial Accuracy")
         
plt.title("Changes to Accuracy Based on Mutation Range")
plt.xlabel("Mutation Range Centroid")
plt.ylabel("RAILS Accuracy")
plt.legend()
outpath = f"results/mutation_range.png"
plt.savefig(outpath, bbox_inches="tight")
```

```python
result_rows = []

G_TYPES = ["crossover","mutate"]
N_NEIGHBORS = 20
N_POPULATION = 4

for operator in G_TYPES:
    CIFAR_CONFIGS = {
    "start_layer": 1,
    "n_class": 10,
    "aise_params": [
        {"hidden_layer": 2, 
         "n_neighbors" : N_NEIGHBORS, 
         "n_population" : N_POPULATION * 10 * N_NEIGHBORS, # variable 
         "sampling_temperature": 1, 
         "max_generation": 10,
         "genop_type": operator,
         "mut_range": (.005, .015)}
    ]
}
    
    print(f"trainning with: genop_type {operator} ")
    rails = RAILS(model, CIFAR_CONFIGS, x_train, y_train)
    
    pred_rails = rails.predict(x_batch.to(DEVICE)).argmax(axis=1)
    acc_rails = (pred_rails == y_batch.numpy()).astype("float").mean().item()
    
    pred_adv_rails = rails.predict(x_adv.to(DEVICE)).argmax(axis=1)
    adv_acc_rails = (pred_adv_rails == y_batch.numpy()).astype("float").mean().item()
    
    row = {
        'RAILS_clean_acc' : acc_rails,
        'RAILS_adversarial_acc' : adv_acc_rails,
        'n_neighbors' :N_NEIGHBORS,
        'n_population' : N_POPULATION * 10 * N_NEIGHBORS,
        'genop_type' : operator,
    }
    
    result_rows.append(row)
    
results = pd.DataFrame(result_rows)
results
```
